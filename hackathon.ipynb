```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import numpy as np
import re
```

```python
sentences = list()
for l in re.split(r"\.\s|\?\s|\!\s|\n",desc):
    if l:
        sentences.append(l)
```

```python
cvec = CountVectorizer(stop_words='english', min_df=3, max_df=0.5, ngram_range=(1,1))
sf = cvec.fit_transform(sentences)
```

```python
transformer = TfidfTransformer()
transformed_weights = transformer.fit_transform(sf)
weights = np.asarray(transformed_weights.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': cvec.get_feature_names(), 'weight': weights})
```

```python
weights_df.sort_values(by='weight', ascending=False).head(10)
```

```python
score =weights_df.sort_values(by='weight', ascending=False)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
K=4

# Select random observation as centroids
Centroids = (X.sample(n=K))
ax.scatter(X_train.a_str_16, X_train.sentenceScore, X_train.shortSentenceScore, c='red', marker='o')
ax.scatter(Centroids["a_str_16"], Centroids["sentenceScore"],Centroids["shortSentenceScore"], c='blue', marker='o')
plt.show()
```

```python
print(Centroids)
```

```python
diff = 1
j=0
while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["a_str_16"]-row_d["a_str_16"])**2
            d2=(row_c["sentenceScore"]-row_d["sentenceScore"])**2
            d3=(row_c["shortSentenceScore"]-row_d["shortSentenceScore"])**2
            d=np.sqrt(d1+d2+d3)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["shortSentenceScore","sentenceScore","a_str_16"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['sentenceScore'] - Centroids['sentenceScore']).sum() + (Centroids_new['a_str_16'] - Centroids['a_str_16']).sum()++ (Centroids_new['shortSentenceScore'] - Centroids['shortSentenceScore']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["sentenceScore","a_str_16","shortSentenceScore"]]
```

```python
color=['blue','green','yellow','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["a_str_16"],data["sentenceScore"],c=color[k])
plt.scatter(Centroids["a_str_16"],Centroids["sentenceScore"],c='red')
plt.xlabel('a_str_16')
plt.ylabel('sentenceScore')
plt.show()
```
